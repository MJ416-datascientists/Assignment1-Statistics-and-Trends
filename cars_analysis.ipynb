Import Libraries

```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns
```

Load the dataset

```python
data = pd.read_csv('Cars dataset.csv')
```

1. Descriptive Statistics

```python
def descriptive_statistics(df):
    """
    Function to get descriptive statistics of the dataset.
    """
    stats = df.describe()
    print("Descriptive Statistics:\n", stats)
    return stats
```

2. Correlation Matrix Heatmap

```python
def plot_correlation_matrix(df):
    """
    Function to plot correlation matrix heatmap to explore relationships between numeric attributes.
    """
    numeric_df = df.select_dtypes(include=['float64', 'int64'])  # Filter numeric columns
    plt.figure(figsize=(10, 8))
    correlation = numeric_df.corr()
    sns.heatmap(correlation, annot=True, cmap="coolwarm", square=True, fmt=".2f")
    plt.title("Correlation Matrix Heatmap")
    plt.show()
```

3. Categorical Plot - Bar Chart (e.g., average selling price by car brand)

```python
def plot_bar_chart(df, category_column, value_column):
    """
    Function to plot a bar chart for a given category and value.
    """
    plt.figure(figsize=(12, 6))
    avg_price_per_category = df.groupby(category_column)[value_column].mean().reset_index()
    sns.barplot(x=category_column, y=value_column, data=avg_price_per_category)
    plt.title(f"Average {value_column} by {category_column}")
    plt.xlabel(category_column)
    plt.ylabel(f"Average {value_column}")
    plt.xticks(rotation=45)
    plt.show()
```

4. Scatter Plot (e.g., relationship between 'km_driven' and 'selling_price')



```python
def plot_scatter(df, x_column, y_column):
    """
    Function to plot a scatter plot between two columns.
    """
    plt.figure(figsize=(10, 6))
    sns.scatterplot(x=x_column, y=y_column, data=df)
    plt.title(f"Scatter Plot of {y_column} vs {x_column}")
    plt.xlabel(x_column)
    plt.ylabel(y_column)
    plt.show()
```

```python
if __name__ == "__main__":
    stats = descriptive_statistics(data)
    plot_correlation_matrix(data)
    plot_bar_chart(data, 'brand', 'selling_price')
    plot_scatter(data, 'km_driven', 'selling_price')
```
